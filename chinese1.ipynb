```python
#coding:utf-8
import os
file_path = "/home/night/data/chinese1/train"
file_names = os.listdir(file_path)
class_dict = {}
for n in file_names:
    img_chars = n.split("_")[1]
    img_chars = img_chars.split(".")[0]
    img_chars = unicode(img_chars,"utf-8")
    for c in img_chars:
        if c in class_dict.keys():
            tmp_num = class_dict[c]
            class_dict[c] = tmp_num + 1
        else:
            class_dict[c] = 1
print class_dict
```
